```python
# Sales Forecast with Linear Regression

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Loading the data
df = pd.read_csv('sales.csv')
X = df[['month']]
y = df['sales']

# 2. Creating and training the model
model = LinearRegression()
model.fit(X, y)

# 3. Making predictions
df['predicted_sales'] = model.predict(X)

# 4. Evaluating the model
mse = mean_squared_error(y, df['predicted_sales'])
r2 = r2_score(y, df['predicted_sales'])

print(f"MSE: {mse:.2f}")
print(f"R²: {r2:.2f}")

# 5. Visualizing the result
plt.figure(figsize=(10, 5))
plt.scatter(df['month'], df['sales'], label='Actual Sales')
plt.plot(df['month'], df['predicted_sales'], color='red', label='Prediction')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales Forecast with Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
