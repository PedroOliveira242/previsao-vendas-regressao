```python
# Previsão de Vendas com Regressão Linear

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1. Carregando os dados
df = pd.read_csv('vendas.csv')
X = df[['mes']]
y = df['vendas']

# 2. Criando e treinando o modelo
modelo = LinearRegression()
modelo.fit(X, y)

# 3. Fazendo previsões
df['vendas_previstas'] = modelo.predict(X)

# 4. Avaliando o modelo
mse = mean_squared_error(y, df['vendas_previstas'])
r2 = r2_score(y, df['vendas_previstas'])

print(f"MSE: {mse:.2f}")
print(f"R²: {r2:.2f}")

# 5. Visualizando o resultado
plt.figure(figsize=(10, 5))
plt.scatter(df['mes'], df['vendas'], label='Vendas Reais')
plt.plot(df['mes'], df['vendas_previstas'], color='red', label='Previsão')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.title('Previsão de Vendas com Regressão Linear')
plt.legend()
plt.grid(True)
plt.show()
```
